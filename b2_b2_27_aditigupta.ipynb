```python
text = "apple banana"
words = text.split()

freq = {}
for word in words:
    freq[word] = freq.get(word, 0) + 1

print(freq)
```

```python
dict1 = {'a': 10, 'b': 20}
dict2 = {'b': 30, 'c': 40}

merged = dict1.copy()
for key, value in dict2.items():
    merged[key] = merged.get(key, 0) + value

print(merged)
```

```python
t = (1, 2, 3, 3, 3)

count_dict = {}
for item in t:
    count_dict[item] = t.count(item)

print(count_dict)
```
