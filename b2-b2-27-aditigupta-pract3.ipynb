```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Housing.csv")

print(df.head())
print(df.info())
print(df.describe())
```

```python
plt.figure()
df['furnishingstatus'].value_counts().plot(kind='bar', color=['skyblue','orange','green'])
plt.title("Count of Furnishing Status")
plt.xlabel("Furnishing Status")
plt.ylabel("Count")
plt.show()
```

```python
counts = df['furnishingstatus'].value_counts()

plt.figure()
plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=90)
plt.title("furnishingstatus Access Distribution")
plt.show()
```

```python
counts = df['mainroad'].value_counts()

plt.figure()
plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=90)
plt.title("Main Road Access Distribution")
plt.show()
```

```python
counts = df['guestroom'].value_counts()

plt.figure()
plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=90)
plt.title("guestroom Access Distribution")
plt.show()
```

```python
counts = df['prefarea'].value_counts()

plt.figure()
plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=90)
plt.title("prefarea Access Distribution")
plt.show()
```

```python

counts = df['airconditioning'].value_counts()

plt.figure()
plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=90)
plt.title("airconditioning Access Distribution")
plt.show()
```

```python


counts = df['hotwaterheating'].value_counts()

plt.figure()
plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=90)
plt.title("hotwaterheating Access Distribution")
plt.show()
```

```python
df_sorted = df.sort_values("price")

plt.plot(df_sorted['price'])
plt.title("Housing Price Trend (Sorted)")
plt.xlabel("Index")
plt.ylabel("Price")
plt.show()
```

```python

plt.plot(df['area'], df['price'])
plt.xlabel("Area (sq ft)")
plt.ylabel("Price")
plt.title("Price vs Area Line Plot")
plt.show()
```

```python
plt.figure()
sns.scatterplot(x='area', y='price', hue='furnishingstatus', data=df)
plt.title("Area vs Price")
plt.show()
```

```python
plt.figure()
sns.boxplot(x=df['price'])
plt.title("Boxplot of Price (Outlier Detection)")
plt.show()
```

```python
plt.figure(figsize=(8,6))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.pairplot(df[['price','area','bedrooms','bathrooms']])
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(9,9))

sns.scatterplot(ax=axes[0,0], x='area', y='price', data=df)
axes[0,0].set_title("Area vs Price")

sns.boxplot(ax=axes[0,1], x='price', data=df)
axes[0,1].set_title("Price Outliers")

df['furnishingstatus'].value_counts().plot(kind='bar', ax=axes[1,0])
axes[1,0].set_title("Furnishing Status Count")

sns.heatmap(df.corr(numeric_only=True), ax=axes[1,1], cmap='coolwarm')
axes[1,1].set_title("Correlation Heatmap")

plt.tight_layout()
plt.show()
```

```python
from scipy import stats

z = np.abs(stats.zscore(df['price']))
df_outliers = df[z > 3]

print("Outliers based on Price:")
print(df_outliers)
```
